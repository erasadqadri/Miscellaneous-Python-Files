# Write a program to reverse the given string (alphabets):

### We can do this using following ways.
1: Manually (Using Indexing)
2: Reverse Function
3: Slicing

#### 1. Manually

```python
string = "mango"
output = ""
i = len(string) - 1  #i = 5-1 = 4
while i>=0:
    output = output + string[i]
    i = i - 1
print(output)
```

```python
string = input("Enter String Here: ")
output = ""
i = len(string)-1
while i>=0:
    output = output + string[i]
    i = i -1
print(output)
```

#### 2. Reverse Function

```python
string = 'Mango'
r = reversed(string)
for n in r:
    print(n)
```

```python
string = "Mango"
r = reversed(string)
output = ''.join(r)
print(output)
```

#### 3. Slicing

```python
string = 'Mango'
#slicing syntax => variable[begin:end:step]
output = string[::-1]
print(output)
```

# Write a program to reverse every alternate word:

```python
string = 'one two three four five six'
l = string.split()
l1 = []
index = 0
while index < len(l):  #len(l) => 6
    if index%2==0:
        l1.append(l[index])
    else:
        l1.append(l[index][::-1])
    index = index+1
#print(l1)
output = ' '.join(l1)
print(output)
```

# Write a program to sort characters of the string, first alphabet symbols followed by digits:

###### input => B4A1D3
###### expected output => ABD134

```python
string = 'B4A1D3'
alphabets = []
digits = []
for characters in string:
    if characters.isalpha():
        alphabets.append(characters)
    
    else:
        digits.append(characters)
#print(digits)
#print(alphabets)
output = ''.join(sorted(alphabets)+sorted(digits))
print(output)
```

```python
string = 'B4A1D3'
alphabets = ''
digits = ''
for characters in string:
    if characters.isalpha():
        alphabets += characters
    else:
        digits += characters
output = ''
for characters in sorted(alphabets):
    output = output+characters
for characters in sorted(digits):
        output = output+characters
print(output)
```

## Write a program to find the Nth term in the following series:

#### 1,1,2,3,4,9,8,27,16,81,32,243,64,729,128,2187........
    for this we will use 2 variables a and b for even and odd index positions

```python
def nseries(n):
    n.self = n
    for(i=1,i<=n,i++):
        if(i%2!=0):
            if(i==1):
                a=1
            else:
                a = a*2
        else:
            if(i==2):
                b = 1
            else:
                b = b*3

output = nseries(11)
print(output)
```

# Wonderbiz Sequence:

```python
def sequence(x,y):
    record = []
    while True:
        yield x
        record.append(x)
        
        for _ in range(record.pop(0)):
            yield y
            record.append(y)

for n in sequence(2,3):
    print(n)
```

```python
count = 1

def doThis():
    global count       # global => global variable
    for i in (1, 2, 7):
        count += 1

doThis() 

print (count) 
#Explanation: The variable "count" declared outside the function is global variable & also the count variable being referenced
#in the function is the same global variable defined outside of the function. So, the changes made to variable in the function
#is reflected to the original variable. So, the output of the program is 4.
```

```python
values = [1, 2, 3, 4] 
numbers = set(values) 

def checknums(num):
    if num in numbers:
        return True
    else:
        return False

for i in filter(checknums, values):
    print(i) 
#Explanation: The function “filter” will return all items from list "values" which return True when passed to the function
#“checknums”. “checknums” will check if the value is in the set. Since all the numbers in the set come from the values list,
#all of the original values in the list will return True.
```

```python
counter = {} 

def addToCounter(country): 
	if country in counter: 
		counter[country] += 1
	else: 
		counter[country] = 1

addToCounter('China') 
addToCounter('Japan') 
addToCounter('china') 

print(len(counter))
print(counter)
#Explanation: The task of “len” function is to return number of keys in a dictionary. Here 3 keys are added to the dictionary
#“country” using the “addToCounter” function.
#Please note carefully – The keys to a dictionary are case sensitive.
```

```python
class Geeks: 
	def __init__(self, id): 
		self.id = id

manager = Geeks(100) 

manager.__dict__['life'] = 49

print (manager.life + len(manager.__dict__)) 
#Explanation : In the above program we are creating a member variable having name ‘life’ by adding it directly to the
#dictionary of the object ‘manager’ of class ‘Geeks’. Total numbers of items in the dictionary is 2, the variables ‘life’ and
#‘id’. Therefore the size or the length of the dictionary is 2 and the variable ‘life’ is assigned a value ’49’. So the sum of
#the variable ‘life’ and the size of the dictionary is 49 + 2 = 51.

print(manager.life)
print(manager.__dict__)
print(len(manager.__dict__))
```
